## Problem: 7 
Create a function which creates an $n×n$ array with $(i , j)$ -entry equal to $i+ j$

```python
# Write your code here
def create_array(n):
    initial_array =np.zeros((n,n),dtype='int')
    for i in range(n):
        for j in range(n):
            initial_array[i,j] =i+j

    return initial_array

```

```python
create_array(3)
```

# Problem: 8
Evaluate $cos$ and $sin$ on the interval $[0,1]$  with 1000 points and then stack the results into a tall array with rows being the $(cos(x),sin(x))$ entries.

```python
# Write your code here
array_x = np.linspace(-1,1,1000)
#print(f"array_x {array_x}")
cos_array = np.cos(array_x)
#print(f"cos_array {cos_array}")
sin_array = np.sin(array_x)
#print(f"sin_array {sin_array}")
stack_array = np.stack((cos_array,sin_array),axis=1 )
print(f"stack_array {stack_array}")
```
